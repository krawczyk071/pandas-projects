```python
import pandas as pd
import numpy as np


df = pd.read_csv('./toexpertlist2.csv',delimiter=';')
```

```python

def to5(str:str,cols=6):
    comas=cols-1
    coma_cnt = str.count(',')
    return ','*(comas-coma_cnt) + str

```

```python
df2 = pd.DataFrame()
df2['address'] = df['address'].apply(to5)
df2
```

```python
df2 = df2['address'].str.split(',', expand=True)
# df
```

```python
df=df.join(df2)
```

```python

df.sort_values([*range(5)],ascending=False,inplace=True)
```
