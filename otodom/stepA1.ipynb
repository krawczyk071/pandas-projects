```python
import pandas as pd
import numpy as np

df = pd.read_csv('./out3.csv',delimiter=';',encoding='utf-8',encoding_errors='replace')
```

```python
df
```

```python
colset1= ['link', 'name', 'sqm','price', 'rooms', 'suburb','neighbourhood',  'type']
df2=df[colset1].sort_values('suburb')
```

```python
# 1 odfiltrowac te bez ceny
df2.price.replace(-1,np.nan,inplace=True)
df2 = df2[df2.price.notna()]
```

```python
ursus = df2[df2.suburb=='Ursus'].sort_values('sqm')
```

```python
pd.set_option('display.max_colwidth', None)
ursus['ppm']=round((ursus.price/ursus.sqm)/1000,2)
ursus[ursus.sqm<70].sort_values('ppm')
```
